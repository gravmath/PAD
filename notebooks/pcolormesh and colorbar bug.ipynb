```python
def cbar_position(current_ax,offset,cbar_width):
    #get the tuple describing the lower x,y corner position
    #and the width and height
    x_pos, y_pos, width, height = current_ax.get_position().bounds
    return [x_pos + width + offset, y_pos, cbar_width, height]  
```

```python
A = np.array([[ 2.,  2.,  3.,  4.],
       [ 2.,  0.,  8.,  5.],
       [ 6.,  4.,  7.,  1.],
       [ 0.,  9.,  9.,  2.],
       [ 9.,  0.,  2.,  9.]])
```

```python
fig = plt.figure(figsize=(10,10))
cb_off = 0.01
cb_wth = 0.01
#first subplot
ax0   = fig.add_subplot(321)
cbax0 = fig.add_axes(cbar_position(ax0,cb_off,cb_wth))
pcm0  = ax0.pcolormesh(A,cmap=cmap.jet,vmin=0,vmax=9)
fig.colorbar(pcm0,cax=cbax0)
#second subplot
ax1   = fig.add_subplot(322)
cbax1 = fig.add_axes(cbar_position(ax1,cb_off,cb_wth))
pcm1  = ax1.pcolormesh(A,cmap=cmap.jet,vmin=1,vmax=8)
fig.colorbar(pcm1,cax=cbax1)
#third subplot
ax2   = fig.add_subplot(323)
cbax2 = fig.add_axes(cbar_position(ax2,cb_off,cb_wth))
pcm2  = ax2.pcolormesh(A,cmap=cmap.jet,vmin=2,vmax=7)
fig.colorbar(pcm2,cax=cbax2)
#fourth subplot
ax3   = fig.add_subplot(324)
cbax3 = fig.add_axes(cbar_position(ax3,cb_off,cb_wth))
pcm3  = ax3.pcolormesh(A,cmap=cmap.jet,vmin=3,vmax=6)
fig.colorbar(pcm3,cax=cbax3)
#fifth subplot
ax4   = fig.add_subplot(325)
cbax4 = fig.add_axes(cbar_position(ax4,cb_off,cb_wth))
pcm4  = ax4.pcolormesh(A,cmap=cmap.jet,vmin=4,vmax=5)
fig.colorbar(pcm4,cax=cbax4)
#sixth subplot
ax5    = fig.add_subplot(326)
cbax_5 = fig.add_axes(cbar_position(ax5,cb_off,cb_wth))
pcm5   = ax5.pcolormesh(A,cmap=cmap.jet,vmin=4.5,vmax=4.5)
fig.colorbar(pcm5,cax=cbax_5)
```
